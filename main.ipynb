```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_train=pd.read_csv("Data/train.csv") #train_pd
df_test=pd.read_csv("Data/test.csv")    #test_pd
```

```python
df_train
```

```python
y_train=df_train['Survived']
y_train
```

```python
df_train=df_train.drop(["Survived"],axis=1)   
df_train
```

```python
train_idx=df_train["PassengerId"]
test_idx=df_test["PassengerId"]

# combining the dataset and reseting the indec
combined_df= pd.concat([df_train,df_test]).reset_index(drop=True)
combined_df
```

```python
sns.heatmap(combined_df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
combined_df.isnull().sum()
```

```python
y=combined_df['Age'].value_counts()
x=y.index.values

plt.figure(figsize=(10,5))
plt.bar(x,y)
plt.title('All Ages')
plt.show()
```

```python
# finding the distribution of Age per class
class_age=combined_df[['Pclass','Age']]

for i in range(1,4):
    plt.figure(figsize=(10,5))
    plt.title('Class '+str(i))
    y=class_age[class_age['Pclass']==i]['Age'].value_counts()
    x=y.index.values

    plt.bar(x,y)
    plt.show()
```

```python
# median age per pclass
median= class_age.groupby('Pclass').median()
median
```

```python
for i in range(3):
    idx = np.where((combined_df['Pclass']==i+1) & (combined_df['Age'].isnull()))[0]
    print("na value in class ",i+1," = ", len(idx))
    combined_df.loc[idx,'Age']= median.values[i][0]
```

```python
combined_df
```

```python
display(combined_df[combined_df['Fare'].isnull()])

fare_s=combined_df[(combined_df['Pclass']==3)&(combined_df['Embarked']=="S")]['Fare']
print('Median', fare_s.median())

combined_df["Fare"].fillna(fare_s.median(),inplace=True)
```

```python
display(combined_df[combined_df['Embarked'].isnull()])

embark_s=combined_df[(combined_df['Pclass']==1)&(combined_df['Fare']>=70)&(combined_df['Fare']<=90)]['Embarked']
print(embark_s.value_counts())
```

```python
combined_df['Embarked'].fillna('c',inplace=True)
```

```python
combined_df['Cabin'].fillna("M",inplace=True)
combined_df['Cabin']=combined_df['Cabin'].str[0]

print(combined_df['Cabin'].value_counts())
# Removing 'T' to avoid overfitting
idx=np.where(combined_df['Cabin']=='T')[0]
combined_df.loc[idx,'Cabin']='M'
```
